##### Identify outliers:
To identify outliers, we can use a commonly used method called the Tukey's fences method. In this method, any data point outside the range of (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR) is considered an outlier, where Q1 is the first quartile and Q3 is the third quartile, and IQR is the interquartile range.

```python
Q1 = q1_df['Measure_X'].quantile(0.25)
Q3 = q1_df['Measure_X'].quantile(0.75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

outliers = q1_df[(q1_df['Measure_X'] < lower_fence) | (q1_df['Measure_X'] > upper_fence)]
print('Outliers:')
print(outliers)
```

### Q4. AT&T was running commercials in 1990 aimed at luring back customers who had switched to one of the other long-distance phone service providers. One such commercial shows a businessman trying to reach Phoenix and mistakenly getting Fiji, where a half-naked native on a beach responds incomprehensibly in Polynesian. When asked about this advertisement, AT&T admitted that the portrayed incident did not actually take place but added that this was an enactment of something that “could happen.” Suppose that one in 200 long-distance telephone calls is misdirected. What is the probability that at least one in five attempted telephone calls reaches the wrong number? (Assume independence of attempts.)

#### Ans: 
    

```python
from math import comb

# Set the parameters
p = 1 / 200  # Probability of success (call being misdirected)
n = 5  # Number of trials (attempted telephone calls)

# Calculate the probability using the complement (1 - P(no successes))
probability = 1 - (1 - p) ** n

# Print the result
print(f"The probability that at least one in five attempted telephone calls reaches the wrong number: {probability}")
```

### Q4.	Returns on a certain business venture, to the nearest $1,000, are known to follow the following probability distribution

                                    x	     P(x)
                                    
                                    -2,000	 0.1
                                    -1,000	 0.1
                                    0	      0.2
                                    1000	   0.2
                                    2000	   0.3
                                    3000	   0.1
                                    
                                    
(i)	What is the most likely monetary outcome of the business venture?

(ii)	Is the venture likely to be successful? Explain.

(iii)	What is the long-term average earning of business ventures of this kind? Explain.

(iv)	What is the good measure of the risk involved in a venture of this kind? Compute this measure



##### (i)	What is the most likely monetary outcome of the business venture?

###### Answer :
The most likely monetary outcome of the business venture can be determined by finding the value with the highest probability (P(x)). In this case, we observe that the value 2000 has the highest probability of 0.3. Therefore, the most likely monetary outcome of the business venture is $2000.

##### (ii) Is the venture likely to be successful? Explain

###### Answer :
To determine if the venture is likely to be successful, we need to consider the probability of earning a positive return (i.e., a return greater than zero). From the given probability distribution, we can see that the probabilities associated with positive returns (1000, 2000, and 3000) add up to 0.2 + 0.3 + 0.1 = 0.6. This means there is a 60% chance of earning a positive return. Based on this information, the venture is likely to be successful.



##### (iii)What is the long-term average earning of business ventures of this kind? Explain

###### Answer :
The long-term average earning of business ventures of this kind can be calculated by multiplying each outcome (x) by its corresponding probability (P(x)), and then summing up these values. Let's calculate it:

```python
outcomes = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

average_earning = sum(x * p for x, p in zip(outcomes, probabilities))


# print('The long-term average earning of business ventures of this kind is $800.0')

print(f"The long-term average earning of business ventures of this kind: ${average_earning}")
```

##### (iv)What is the good measure of the risk involved in a venture of this kind? Compute this measure

###### Answer :
A common measure of risk in a venture of this kind is the standard deviation. The standard deviation measures the dispersion or variability of the returns. Let's calculate the standard deviation using the given probabilities:

```python
import math

mean = sum(x * p for x, p in zip(outcomes, probabilities))
variance = sum((x - mean) ** 2 * p for x, p in zip(outcomes, probabilities))
std_deviation = math.sqrt(variance)

print(f"The standard deviation of the venture's returns: ${std_deviation:.2f}")
```
